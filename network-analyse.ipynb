مجموعه‌داده انتخابی:

شبکه ساخته شده با مدل Barabasi-Albert شامل 1500 گره است که برای این تحلیل مناسب است.

نمایش خروجی کد:

خروجی شامل مقادیر محاسبه شده برای ویژگی‌های شبکه و نمودارهای رسم شده است.

````python
import networkx as nx
import matplotlib.pyplot as plt

# 1. ساخت شبکه با تعداد گره بیشتر از 1000
G = nx.barabasi_albert_graph(1500, 5)  # یک شبکه با مدل ترجیحی اتصال

# 2. فرمول های محاسبه ویژگی‌های شبکه
'''این مقدار نشان‌دهنده بیشترین تعداد اتصالاتی است که یک گره در شبکه دارد.
 درجه یک گره برابر است با تعداد یال‌هایی که به آن گره متصل هستند.
 گره‌هایی با درجه بالا معمولاً نقش کلیدی در شبکه دارند
 و به عنوان "گره‌های حیاتی" یا "هاب" شناخته می‌شوند.'''

max_degree = max(dict(G.degree()).values())  # ماکزیمم درجه
'''این مقدار میانگین تعداد
یال‌های متصل به تمام گره‌های شبکه است و از فرمول زیر محاسبه می‌شود
: $$ \text{Average Degree} = \frac{\sum_{i=1}^{N} k_i}{N} $$ که در آن
𝑘
𝑖
 درجه گره
𝑖
 و
𝑁
 تعداد کل گره‌ها است.'''
avg_degree = sum(dict(G.degree()).values()) / G.number_of_nodes()  # متوسط درجه
'''این مقدار میانگین طول
کوتاه‌ترین مسیرها بین تمام جفت‌های گره‌ها در شبکه است.
 این ویژگی نشان‌دهنده "کوچکی دنیا" یا ارتباط کلی بین گره‌ها در شبکه است'''
avg_distance = nx.average_shortest_path_length(G)  # متوسط فاصله
'''قطر شبکه طولانی‌ترین مسیر در میان کوتاه‌ترین مسیرها بین هر دو گره است.
 این مفهوم نشان‌دهنده حداکثر فاصله بین گره‌های شبکه است.'''
diameter = nx.diameter(G)  # قطر
'''ضریب خوشه‌بندی نشان‌دهنده میزان خوشه‌بندی گره‌ها در شبکه است و از فرمول زیر محاسبه می‌شود:
 $$ C_i = \frac{2 \times \text{number of triangles involving } i}{\text{number of connected triples through } i} $$ متوسط ضریب خوشه‌بندی برابر میانگین مقادیر خوشه‌بندی برای تمام گره‌های
شبکه است. این ویژگی نشان می‌دهد که گره‌های شبکه چقدر تمایل به تشکیل گروه‌های سه‌گانه (مثلث‌ها) دارند.'''
clustering_coefficient = nx.average_clustering(G)  # متوسط ضریب خوشه‌بندی

# چاپ نتایج
print("ماکزیمم درجه:", max_degree)
print("متوسط درجه:", avg_degree)
print("متوسط فاصله:", avg_distance)
print("قطر شبکه:", diameter)
print("متوسط ضریب خوشه‌بندی:", clustering_coefficient)

# 3. رسم نمودار توزیع درجه
'''توزیع عادی (Degree Distribution): این نمودار تعداد گره‌هایی با هر درجه مشخص را نشان می‌دهد. معمولاً به صورت یک هیستوگرام رسم می‌شود.

توزیع log-log: نسخه‌ای از نمودار توزیع درجه است که محورهای آن به صورت لگاریتمی
نمایش داده می‌شوند.
این نوع نمودار برای نمایش رفتار نمایی یا مقیاس‌پذیری شبکه‌ها استفاده می‌شود.'''
degrees = [degree for node, degree in G.degree()]
plt.figure(figsize=(12, 6))

# توزیع عادی
plt.subplot(1, 2, 1)
plt.hist(degrees, bins=30, color='blue', edgecolor='black')
plt.title("توزیع درجه (عادی)")
plt.xlabel("درجه")
plt.ylabel("تعداد")

# توزیع log-log
plt.subplot(1, 2, 2)
plt.hist(degrees, bins=30, color='green', edgecolor='black', log=True)
plt.title("توزیع درجه (log-log)")
plt.xlabel("درجه")
plt.ylabel("تعداد (log)")

plt.tight_layout()
plt.show()

# 4. محاسبات برای 5 گره با بالاترین درجه
top_5_nodes = sorted(G.degree, key=lambda x: x[1], reverse=True)[:5]  # گره‌های با بیشترین درجه
for node, degree in top_5_nodes:
    clustering = nx.clustering(G, node)
    print(f"گره {node} | درجه: {degree} | ضریب خوشه‌بندی: {clustering}")
````

تحلیل نتایج و نمودارهای ارائه‌شده:

### ویژگی‌های شبکه:
1. **ماکزیمم درجه: 157**
   - این عدد نشان می‌دهد که پراتصال‌ترین گره شبکه (گره 8) دارای 157 ارتباط با سایر گره‌ها است. گره‌های با درجه بالا اغلب نقش کلیدی در شبکه دارند و می‌توانند "هاب" باشند که بر ساختار و عملکرد کلی شبکه تأثیرگذار هستند.

2. **متوسط درجه: 9.97**
   - متوسط درجه نشان می‌دهد که به طور میانگین هر گره تقریباً 10 ارتباط دارد. این عدد به ما ایده‌ای از چگالی شبکه می‌دهد؛ شبکه شما نسبتاً پراکنده است.

3. **متوسط فاصله: 3.11**
   - این مقدار نشان می‌دهد که به طور میانگین کوتاه‌ترین مسیر بین هر دو گره تنها شامل سه گره است. این ویژگی نشان‌دهنده خاصیت "جهان کوچک" در شبکه است که در بسیاری از شبکه‌های دنیای واقعی مانند شبکه‌های اجتماعی و زیستی یافت می‌شود.

4. **قطر شبکه: 5**
   - قطر نشان‌دهنده بیشترین فاصله بین دو گره در شبکه است. کوچک بودن قطر (عدد 5) نشان‌دهنده ساختار متصل و کارآمد شبکه است.

5. **متوسط ضریب خوشه‌بندی: 0.027**
   - ضریب خوشه‌بندی پایین نشان می‌دهد که گره‌ها تمایل کمی به تشکیل گروه‌های محلی یا خوشه دارند. این ممکن است نشان‌دهنده ساختار غیرمحلی شبکه باشد، جایی که اتصالات بیشتر بین گروه‌های مختلف پخش شده‌اند.

### تحلیل 5 گره با بالاترین درجه:
- **گره 8:** با بالاترین درجه (157) ضریب خوشه‌بندی 0.018 دارد. این نشان می‌دهد که این گره با تعداد زیادی گره دیگر متصل است، اما این اتصالات به ندرت مثلث تشکیل داده‌اند.
- **گره 6 و گره 7:** این گره‌ها نیز اتصالات بسیاری دارند (به ترتیب 134 و 107 درجه) و ضریب خوشه‌بندی آنها نیز پایین است. این رفتار متناسب با گره‌هایی است که به عنوان هاب عمل می‌کنند و اغلب شبکه‌های مقیاس‌پذیر این خصوصیت را نشان می‌دهند.
- **گره 0 و گره 12:** با درجات 102 و 99، این گره‌ها نیز نقش مهمی در شبکه دارند، اما کمی ضریب خوشه‌بندی بالاتری نسبت به گره‌های دیگر دارند. این نشان‌دهنده تمایل کمی بیشتر به تشکیل خوشه‌های محلی است.

---

### تحلیل نمودارهای توزیع درجه:
1. **توزیع عادی (نمودار چپ):**
   - اکثر گره‌ها درجه‌های پایینی دارند و تعداد کمی از گره‌ها درجه‌های بالایی دارند. این نشان‌دهنده توزیع "مقیاس‌پذیر" است که در شبکه‌های واقعی مانند شبکه‌های اجتماعی و بیولوژیکی دیده می‌شود.

2. **توزیع log-log (نمودار راست):**
   - نمایش لگاریتمی توزیع نشان می‌دهد که گره‌های با درجه بالا (هاب‌ها) تعداد کمی دارند، در حالی که بیشتر گره‌ها درجه‌های پایینی دارند. این خصوصیت تایید می‌کند که شبکه از نوع "رابطه ترجیحی" است.

---

این تحلیل‌ها نشان‌دهنده یک شبکه مقیاس‌پذیر و جهان کوچک است که در بسیاری از سیستم‌های واقعی (مانند شبکه‌های اجتماعی) دیده می‌شود.

````python
import networkx as nx
import matplotlib.pyplot as plt

# 1. ساخت شبکه با تعداد گره بیشتر از 1000
G = nx.barabasi_albert_graph(1500, 50)  # یک شبکه با مدل ترجیحی اتصال

# 2. محاسبه ویژگی‌های شبکه
max_degree = max(dict(G.degree()).values())  # ماکزیمم درجه
avg_degree = sum(dict(G.degree()).values()) / G.number_of_nodes()  # متوسط درجه
avg_distance = nx.average_shortest_path_length(G)  # متوسط فاصله
diameter = nx.diameter(G)  # قطر
clustering_coefficient = nx.average_clustering(G)  # متوسط ضریب خوشه‌بندی

# چاپ نتایج
print("ماکزیمم درجه:", max_degree)
print("متوسط درجه:", avg_degree)
print("متوسط فاصله:", avg_distance)
print("قطر شبکه:", diameter)
print("متوسط ضریب خوشه‌بندی:", clustering_coefficient)

# 3. رسم نمودار توزیع درجه
degrees = [degree for node, degree in G.degree()]
plt.figure(figsize=(6, 6))

# توزیع عادی
plt.subplot(1, 2, 1)
plt.hist(degrees, bins=30, color='blue', edgecolor='black')
plt.title("توزیع درجه (عادی)")
plt.xlabel("درجه")
plt.ylabel("تعداد")

# توزیع log-log
plt.subplot(1, 2, 2)
plt.hist(degrees, bins=30, color='green', edgecolor='black', log=True)
plt.title("توزیع درجه (log-log)")
plt.xlabel("درجه")
plt.ylabel("تعداد (log)")

plt.tight_layout()
plt.show()

# 4. محاسبات برای 5 گره با بالاترین درجه
top_5_nodes = sorted(G.degree, key=lambda x: x[1], reverse=True)[:5]  # گره‌های با بیشترین درجه
for node, degree in top_5_nodes:
    clustering = nx.clustering(G, node)
    print(f"گره {node} | درجه: {degree} | ضریب خوشه‌بندی: {clustering}")
````
