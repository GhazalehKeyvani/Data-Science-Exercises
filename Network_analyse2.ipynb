```python
import networkx as nx
import matplotlib.pyplot as plt

# پارامترها
n = 100  # تعداد نودها
p = 0.1  # احتمال ایجاد یال

# تولید گراف
G_er = nx.erdos_renyi_graph(n, p)

# محاسبه پارامترها
avg_degree = sum(dict(G_er.degree()).values()) / n
clustering_coeff = nx.average_clustering(G_er)
avg_path_length = nx.average_shortest_path_length(G_er)

# تحلیل
print(f"میانگین درجه: {avg_degree:.2f}")
print(f"ضریب خوشه‌بندی: {clustering_coeff:.2f}")
print(f"میانگین طول مسیر: {avg_path_length:.2f}")

# رسم گراف
nx.draw(G_er, node_size=50)
plt.title("Erdős-Rényi Graph (n=100, p=0.1)")
plt.show()
```

```python
# پارامترها
n = 100  # تعداد نودها
k = 6    # میانگین درجه
beta = 0.2  # احتمال بازپیوندی

# تولید گراف
G_ws = nx.watts_strogatz_graph(n, k, beta)

# محاسبه پارامترها
clustering_coeff_ws = nx.average_clustering(G_ws)
avg_path_length_ws = nx.average_shortest_path_length(G_ws)

# تحلیل
print(f"ضریب خوشه‌بندی: {clustering_coeff_ws:.2f}")
print(f"میانگین طول مسیر: {avg_path_length_ws:.2f}")

# رسم گراف
nx.draw(G_ws, node_size=50)
plt.title("Watts-Strogatz Graph (n=100, k=6, β=0.2)")
plt.show()
```
